```python
from statistics import correlation

import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
#Heatmap for missingness
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar="False", ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")
```

```python
#Numeric representation of missing books
import numpy as np
books["missing_description"] = np.where(books["description"].isna(), 1, 0)
#Convert year to age of the book
books["age_of_book"] = 2025 - books["published_year"]
```

```python
#List of columns we're going to use
columns_of_interest = ["num_pages", "missing_description", "age_of_book", "average_rating"]
#Correlation matrix of pandas data frame
correlation_matrix = books[columns_of_interest].corr(method="spearman")

#Heatmap of correlation matrix
sns.set_theme(style="white")
plt.figure(figsize=[8,6])
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm",
                      cbar_kws={"label": "Correlation"})
heatmap.set_title("Correlation Heatmap")
plt.show()
```

```python
#How many books with missing values
#Condtional logic in pandas |(or)
books[(books["description"].isna()) |
      (books["num_pages"].isna()) |
      (books["average_rating"].isna()) |
      (books["published_year"].isna())
]
```

```python
#Data frame to delete books with missing values
#Negtion (~) for the conditional logic
book_missing = books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["average_rating"].isna()) &
      ~(books["published_year"].isna())
]
book_missing
```

```python
#Category distribution
#Sort values by count in descending order
book_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
#There is a longtail problem with the categories that need normalisation. This will be done with LLMs (text classification).
```

```python
#Length of descriptions
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
book_missing
```

```python
book_missing.loc[book_missing["words_in_description"].between(1, 4), "description"]
```

```python
book_missing.loc[book_missing["words_in_description"].between(25, 34), "description"]
```

```python
#Removing short descriptions with 25> words
book_missing_25_words = book_missing[book_missing["words_in_description"]>= 25]
book_missing_25_words
```

```python
book_missing_25_words["title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"],  #when no subtitle use title
             book_missing_25_words[["title", "subtitle"]].astype(str).agg(":".join, axis=1))    #when there's both aggregate title and subtitle
)
book_missing_25_words
```

```python
#Tag each description with unique identifier (for the vector database)
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
book_missing_25_words
```

```python
#Final cleaning steps
#Drop columns used to filter and clean
(book_missing.drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
.to_csv("books_cleaned.csv", index=False))
```
